```python
import pandas as pd
```

```python
data = pd.read_csv('health_indicators.csv')
```

```python
df = data
```

```python
county_totals = df.groupby('County').agg({
    'Numerator_Total': 'sum',
    'Denominator_Total': 'sum'
}).reset_index()
```

```python
county_totals
```

```python
county_totals['County_Rate'] = (county_totals['Numerator_Total'] / (county_totals['Denominator_Total']/4))*1000
```

```python
county_totals = county_totals[['County', 'County_Rate']]
county_totals
```

```python
county_totals.to_csv('county_rate.csv', index=False)
```
